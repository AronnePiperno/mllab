# MLP lab using numpy

In this lab I try to create a multi layer perceptron from scratch using Numpy

MLP is a neural network that can be used both for regression and classification, it is formed by an input layer $n$ hidden layers and a output layer.

```python
import numpy as np
```

```python
class MLP:
    # Our MLP needs 3 parameters: how many inputs we have, the size of the hidden layer (the number of nodes per layer) and how many outputs we want
    def __init__(self, input_size, hidden_size, ouput_size):
        # we initialize the weights and the biases vectors
        self.input_weights = np.random.randn(input_size, hidden_size)
        self.input_bias = np.zeros((1, hidden_size))
        self.output_weights = np.random.randn(hidden_size, ouput_size)
        self.output_bias = np.zeros((1, ouput_size))

    def forward(self, input):
        hid = input @ self.input_weights + self.input_bias
        out = hid @ self.output_weights + self.output_bias

        return out

    # def forward(self, input):
    #     # hidden layer computation
    #     hid = np.dot(input, self.input_weights) + self.input_bias


    #     # output layer computation
    #     out = np.dot(hid, self.output_weights) + self.output_bias
    #     return out

    
```

```python
# set a seef for reproducibility
np.random.seed(0)

mlp = MLP(3, 5, 2)
input = np.array([[1, 2, 3]])
output = mlp.forward(input)
print(output)
```
